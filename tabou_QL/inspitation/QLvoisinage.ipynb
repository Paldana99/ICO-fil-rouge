```python
#fonction usuelles pour des fonctions de voisinage
def findVehicule(solution):
    l=len(solution)
    vehicule=[]
    for i in range(1,l-1):
        if solution[i]==0:
            vehicule.append(i)
    return vehicule

def findRoute(solution):
    route=[]
    vehicule=findVehicule(solution)
    for j in range(len(vehicule)):
        if j==0:
            route.append(solution[1:vehicule[j]:])
        else:
            route.append(solution[vehicule[j-1]+1:vehicule[j]:])
    route.append(solution[vehicule[j]+1:-1:])
    return route

def findSolution(route):
    solution=[]
    for i in range(len(route)):
        solution.append(0)
        solution.extend(route[i])
    solution.append(0)
    return solution
```

```python
# 8 fonctions de voisinages
def swapIntra(s0):
    route=findRoute(s0)
    i=rd.randint(0,len(route)-1)
    if len(route[i])>1:
        indice1=rd.choice(range(0,len(route[i])-1))
        indice2=rd.choice(range(indice1+1,len(route[i])))
        route[i][indice1],route[i][indice2]=route[i][indice2],route[i][indice1]
    s_new=findSolution(route)
    return s_new
```

```python
def swapInter(s0):
    route=findRoute(s0)
    i=rd.randint(0,len(route)-1)
    while(1):
        j=rd.randint(0,len(route)-1)
        if i!=j:
            break
    indice1=rd.choice(range(0,len(route[i])))
    indice2=rd.choice(range(0,len(route[j])))
    route[i][indice1],route[j][indice2]=route[j][indice2],route[i][indice1]
    s_new=findSolution(route)
    return s_new
```

```python
def shiftIntra(s0):
    route=findRoute(s0)
    while(1):
        i=rd.randint(0,len(route)-1)
        if len(route[i])>1:
            break
    client=rd.choice(route[i])
    route[i].remove(client)
    indice=rd.choice(range(0,len(route[i])))
    route[i].insert(indice,client)
    s_new=findSolution(route)
    return s_new
```

```python
def shiftInter(s0):
    """shiftInter"""
    route=findRoute(s0)
    while(1):
        i=rd.randint(0,len(route)-1)
        if len(route[i])>1:
            break
    client=rd.choice(route[i])
    route[i].remove(client)
    while(1):
        j=rd.randint(0,len(route)-1)
        if i!=j:
            break
    indice=rd.choice(range(0,len(route[j])))
    route[j].insert(indice,client)
    s_new=findSolution(route)
    return s_new
```

```python
def swapIntra2(s0):
    route=findRoute(s0)
    i=rd.randint(0,len(route)-1)
    if len(route[i])>1:
        indice1=rd.choice(range(0,len(route[i])-1))
        indice2=rd.choice(range(0,len(route[i])-1))
        route[i][indice1],route[i][indice2]=route[i][indice2],route[i][indice1]
        route[i][indice1+1],route[i][indice2+1]=route[i][indice2+1],route[i][indice1+1]
    s_new=findSolution(route)
    return s_new
```

```python
def shiftIntra2(s0):
    route=findRoute(s0)
    while(1):
        i=rd.randint(0,len(route)-1)
        if len(route[i])>2:
            break
    indice1=rd.choice(range(0,len(route[i])-1))
    client1=route[i][indice1]
    client2=route[i][indice1+1]
    route[i].remove(client1)
    route[i].remove(client2)
    indice=rd.choice(range(0,len(route[i])))
    route[i].insert(indice,client1)
    route[i].insert(indice+1,client2)
    s_new=findSolution(route)
    return s_new
```

```python
def eliminePlusPetit(s0):
    route=findRoute(s0)
    longueur=[]
    for i in range(len(route)):
        longueur.append(len(route[i]))
    indice=longueur.index(min(longueur))
    solution=s0[:]
    if indice==0:
        client=route[indice][-1]
        indice1=solution.index(client)
        solution.pop(indice1+1)
    else:
        client=route[indice][0]
        indice1=solution.index(client)
        solution.pop(indice1-1)
    return solution
```

```python
def elimineAlea(s0):
    route=findRoute(s0)
    indice=rd.randint(0,len(route)-1)
    solution=s0[:]
    if indice==0:
        client=route[indice][-1]
        indice1=solution.index(client)
        solution.pop(indice1+1)
    else:
        client=route[indice][0]
        indice1=solution.index(client)
        solution.pop(indice1-1)
    return solution
```

```python
def take_action(st, Q, eps):
    # Take an action
    if rd.uniform(0, 1) < eps:
        #exploration
        action = rd.randint(0, 7)
    else: # Or greedy action
        #exploitation
        action = np.argmax(Q[st])
    return action
```
